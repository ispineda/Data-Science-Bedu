# 1. Identificación del problema

En la actualidad la industria de los videojuegos ha crecido de manera exponencial y a lo largo de los últimos años teniendo una valuación de 159.5 billones de dólares según [Statista](https://www.statista.com/topics/868/video-games/), el mismo sitio afirma que la industria obtiene ganancias de aproximadamente 19.7 billones de dólares por año. Según datos del mismo sitio un 56% de los juegos que se lanzan cada año son desarrollados para PC, esto debido a que es más fácil desarrollarlos para esta plataforma por lo que resulta ser una industria con tanto crecimiento y futuro, y es normal que cada vez más emprendedores quieran desarrollar sus videojuegos y lanzarlos al mercado.


Como se pudo observar en el análisis realizado en el anterior módulo, esta industria año con año se satura de más videojuegos lo que nos permite realizar las siguientes interrogantes: De todos los videojuegos desarrollados ¿Cuántos logran tener éxito? ¿Por qué tienen éxito? y como principales interrogantes ¿Hacia dónde se dirige el mercado? y ¿Cómo se comporta dicho mercado?. 
Debido éstas interrogantes, consideramos que es importante encontrar sus respectivas respuestas ya que como emprendedores, estudios de desarrollo grandes o pequeños necesitan analizar el estado actual del mercado, así como sus posibles tendencias y proyecciones del mismo para lograr el mayor éxito posible.


Los juegos creados de manera independiente suelen ser etiquetados en un género denominado “indie”. De donde los desarrolladores presentan incertidumbre al momento de lanzar sus creaciones al mercado, ya que pueden presentar cuestionamientos referente al mercado. Un ejemplo de ello es Adam Saltsman, desarrollador conocido de juegos independientes, menciona que de la totalidad de los juegos lanzados comercialmente en el género indie la mayoría no rebasan las 3.000 copias en ventas, mencionando inclusive que para que un juego alcance más de las 30.000 copias en ventas se deben considerar factores como: la producción, el tiempo, la inversión, las tendencias en géneros y el costo de venta. Las cuales son características que no todos los desarrolladores pueden estimar en su momento, para lograr un mayor acercamiento a los jugadores.

## 4. Limpieza y transformación de los datos

````
# Limpieza y transofrmación de datos: Amazon
def rename(row):
  return row.split(' - ')[0]

try:
  amazon = amazon.drop(['Unnamed: 0', 'time', 'Rank', 'Reviews', 'Max_prices'], axis=1)
except:
  print('Variables eliminadas')

amazon = amazon.rename({'Product Names': 'PRODUCT_NAME', 'Stars': 'STARS', 'Authors/Company': 'COMPANY', 'Edition/Console': 'CONSOLE', 'Price_std_or_min': 'PRICE'}, axis=1)
amazon.dropna()
amazon['PRODUCT_NAME'] = pd.Series(map(rename, amazon['PRODUCT_NAME'].astype(str)))
amazon.info()
amazon.head()
````

````
# Limpieza y transformación de datos: Apple Store
apple_store = pd.merge(apple_store_1, apple_store_2, left_on='id', right_on='id')
apple_store = apple_store[['track_name_x', 'rating_count_tot_x', 'user_rating_x', 'prime_genre_x']]
apple_store = apple_store.rename({'track_name_x': 'PRODUCT_NAME', 'rating_count_tot_x': 'RATING', 'user_rating_x' : 'USER_RATING', 'prime_genre_x' : 'GENRE'}, axis=1)
apple_store = apple_store[apple_store['GENRE'] == 'Games']
apple_store.info()
apple_store.head()
````

````
# Limpieza y transformación de datos: Consoles
consoles_1 = consoles_1.rename({'ConsoleID': 'CONSOLE', 'Manufacturer' : 'Publisher'}, axis=1)
consoles_2 = consoles_2.rename({'Platform': 'CONSOLE'}, axis=1)
consoles_2 = consoles_2.replace({'CONSOLE':["X"]}, {'CONSOLE': 'XB'}, regex=False)
print(consoles_1.sort_values('Publisher', ascending=False)['Publisher'].unique())
print(consoles_2.sort_values('Publisher', ascending=False)['Publisher'].unique())
consoles = pd.concat([consoles_1, consoles_2])
consoles.info()
consoles.head()
````

````
# Limpieza y transformación de datos: Google Play Store
g_play_store_1.equals(g_play_store_2)
g_play_store_1['Category'].unique()
g_play_store_1 = g_play_store_1[g_play_store_1['Category'] == 'GAME']
g_play_store = g_play_store_1.drop(['Size', 'Content Rating', 'Last Updated', 'Current Ver', 'Android Ver'], axis=1)
g_play_store.info()
g_play_store.head()
````

````
# Limpieza y transformación de datos: Developers
developers = pd.concat([indie_developers, videogames_developers])
developers = developers.drop(['Notes', 'Administrative division', 'Est.', 'Notable games, series or franchises'], axis=1)
developers.info()
developers.head(100)
````

````
# Limpieza y transformación de datos: Microsoft Store
microsoft_store = microsoft_store.drop(['Date'], axis=1)
microsoft_store['Category'].unique()
## Éste dataset no es de utilidad, debido a que no contiene datos relacionados a videojuegos.
````

````
# Limpieza y transformación de datos: PlayStation Store
ps_1['name'] = ps_1['game']
ps_2['name'] = ps_2['Title']
ps_sales['name'] = ps_sales['Game']

ps_1['name'] = ps_1.replace({'name':["'","|","-","/",":"," ",","]}, {'name': ''}, regex = True)['name'].str.upper()
ps_2['name'] = ps_2.replace({'name': ["'","|","-","/",":"," ",","]}, {'name': ''}, regex=True)['name'].str.upper()
ps_sales['name'] = ps_sales.replace({'name': ["'","|","-","/",":"," ",","]}, {'name': ''}, regex=True)['name'].str.upper()

ps = pd.merge(ps_sales, pd.merge(ps_1, ps_2, on='name'), on='name')
ps = ps.drop(['name', 'Title', 'game', 'Unnamed: 0_x', 'comp_perc', 'url', 'min_comp_time', 'max_comp_time', 'Unnamed: 0_y', 'Genre_y', 'Addons', 'Publisher_y'], axis=1)

ps['Release date EU'] = pd.to_datetime(ps['Release date EU'], format='%b %d, %Y', errors='coerce')
ps['Release date JP'] = pd.to_datetime(ps['Release date JP'], format='%b %d, %Y', errors='coerce')
ps['Release date NA'] = pd.to_datetime(ps['Release date NA'], format='%b %d, %Y', errors='coerce')

ps = ps.rename({'Genre_x':'Genre', 'Publisher_x':'Publisher'}, axis=1)

ps.info()
ps.head()
````

````
# Limpieza y transformación de datos: Steam
steam = steam[['name','release_date','owners','positive_ratings','negative_ratings','price']]
steam['release_date'] = pd.to_datetime(steam['release_date'], format='%Y-%m-%d', errors='coerce')
steam.info()
steam.head()
````

````
# Limpieza y transformación de datos: Videogame ratings & scores
videogames_ratings = videogames_ratings.drop(['id','developer','rating'], axis = 1)
videogames_ratings.head()
videogames_ratings.info()

metacritic = metacritic.drop(['metascore'], axis = 1)
metacritic['userscore'] = pd.to_numeric(metacritic['userscore'], errors='coerce')
metacritic['date'] = pd.to_datetime(metacritic['date'], format='%b %d, %Y', errors='coerce')
metacritic.head()
metacritic.info()
````
