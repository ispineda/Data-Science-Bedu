### Ejemplo 4: Scatterplots con variables condicionantes

1. Objetivos:
    - Aprender a generar e interpretar scatterplots con variables condicionantes
 
---
    
2. Desarrollo:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Remoto Datasets/Remoto athlete_olympic_events-clean.csv', index_col=0)

df.head()
```

Por ejemplo, digamos que queremos visualizar las diferencias de peso y altura que hay entre los atletas que juegan basketball, volleyball, football y beisball:

```python
df_grouped = df.groupby('name')[['height', 'weight', 'sport', 'sex']].max()
```

```python
df_ball = df_grouped[df_grouped['sport'].isin(['Basketball', 'Volleyball', 'Football', 'Baseball'])]
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot()
sns.scatterplot(x=df_ball['height'], y=df_ball['weight'], hue=df_ball['sport'], style=df_ball['sport'], ax=ax);
```

Utilizando este scatterplot por categor√≠as ya podemos darnos una idea de las diferencias. Pero hay cierta confusi√≥n en la parte central de la gr√°fica donde est√°n concentrados todos los deportes. Es dif√≠cil diferenciar los puntos cuando hay aglomeraciones tan grandes. Podemos entonces generar un scatterplot con variables condicionantes para ayudarnos:

```python
basket = df_ball[df_ball['sport'] == 'Basketball']
volley = df_ball[df_ball['sport'] == 'Volleyball']
foot = df_ball[df_ball['sport'] == 'Football']
base = df_ball[df_ball['sport'] == 'Baseball']
```

```python
sns.set_style('whitegrid')
palette = ["#223127","#9c0d38","#ce5374","#dbbbf5","#F9A03F"]

fig = plt.figure(figsize=(8, 8))
axes = fig.subplots(2, 2, sharex=True, sharey=True)

sns.scatterplot(x=basket['height'], y=basket['weight'], hue=basket['sex'], ax=axes[0, 0], color=palette[4])
sns.scatterplot(x=volley['height'], y=volley['weight'], ax=axes[0, 1], color=palette[1])
sns.scatterplot(x=foot['height'], y=foot['weight'], ax=axes[1, 0], color=palette[2])
sns.scatterplot(x=base['height'], y=base['weight'], ax=axes[1, 1], color=palette[3]);

axes[0, 0].set_title('Basketball')
axes[0, 1].set_title('Volleyball')
axes[1, 0].set_title('Football')
axes[1, 1].set_title('Baseball')

fig.suptitle('Relationship between height and weight in 4 sports', fontsize=15, c=palette[0], x=0.53)

plt.tight_layout()
plt.subplots_adjust(top=0.9)
```

Otra manera de verlo es mediante Seaborn con los sig parametros

```python
sns.relplot(data=df_ball, x='height', y='weight', kind='scatter', hue='sex',style='sex',col='sport', col_wrap=2);
```

¬°Genial¬° Ahora s√≠ tenemos mucha m√°s claridad acerca de c√≥mo est√°n repartidos los atletas que juegan estos 4 deportes. Podemos ver que cada deporte tiene ciertas tendencias muy claras. Aunque hay mucha superposici√≥n, hay diferencias claras.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Remoto Datasets/Remoto athlete_olympic_events-clean.csv', index_col=0)

df.head()
```

Por ejemplo, mira qu√© pasa cuando queremos visualizar la relaci√≥n entre peso y altura utilizando todos los datos de nuestro dataset:

```python
df_grouped = df.groupby('name')[['height', 'weight']].max()
```

```python
sns.jointplot(x='height', y='weight', data=df_grouped, kind='hex', ratio=5, joint_kws={'gridsize': 50}, marginal_kws={'bins':50});
```

Esta gr√°fica permite que visualicemos con mucho m√°s detalle la distribuci√≥n de nuestros datos. Observa tambi√©n los histogramas que se encuentran en la parte superior y derecha de la gr√°fica. ### Ejemplo 6: Mapas Cloropl√©ticos

1. Objetivos:
    - Aprender a generar e interpretar mapas cloropl√©ticos
 
---
    
2. Desarrollo:

Para generar un mapa cloropl√©tico necesitamos dos fuentes de datos:

1. La primera es un archivo que contenga datos en un formato llamado GeoJSON. El formato GeoJSON utiliza el formato JSON para representar caracter√≠sticas geogr√°ficas de una manera que pueda ser interpretada por una computadora. Un archivo GeoJSON puede incluir puntos en un mapa, regiones geogr√°ficas, anotaciones, nombres de regiones, etc. Este archivo nos sirve para poder colocar objetos sobre un mapa.

2. La segunda es nuestro dataset. Este dataset es como cualquier otro que hemos utilizando anteriormente. Lo √∫nico importante es que tenemos que asegurarnos de que haya una manera de relacionar nuestro dataset con nuestro archivo GeoJSON. Esto se hace de una manera similar a como funcionan los joins de SQL. Relacionando una columna de nuestro dataset con una llave de nuestro GeoJSON podemos entonces indicarle a nuestro programa c√≥mo dibujar cosas sobre un mapa.

Una de las librer√≠as que podemos utilizar para dibujar mapas se llama `folium`. Se utiliza de la siguiente manera:

```python
import pandas as pd
import folium
```

```python
state_geo = f'/content/drive/MyDrive/Remoto Datasets/Remoto us_states.json'
state_unemployment = f'/content/drive/MyDrive/Remoto Datasets/Remoto us_unemployment-oct_2012.csv'
state_data = pd.read_csv(state_unemployment)

m = folium.Map(location=[38, -102], zoom_start=4, width='100%', height='100%')

folium.Choropleth(
    geo_data=state_geo,
    name='choropleth',
    data=state_data,
    columns=['State', 'Unemployment'],
    key_on='feature.id',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Unemployment Rate (%)'
).add_to(m)

folium.LayerControl().add_to(m)

m
```

Es posible agregar popups para poder visualizar los valores de cada regi√≥n. Si te interesa aprender c√≥mo hacer esto, puedes visitar este link: https://vverde.github.io/blob/interactivechoropleth.html.

### Ejemplo 7: Gr√°ficas de barras apiladas

1. Objetivos:
    - Aprender a generar e interpretar gr√°ficas de barras apiladas
 
---
    
2. Desarrollo:

```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Remoto Datasets/Remoto athlete_olympic_events-clean.csv', index_col=0)

df.head()
```

Por ejemplo, digamos que queremos graficar la frecuencia de los 4 deportes que analizamos anteriormente. Queremos ver cu√°ntas veces aparece cada uno en nuestro dataset para compararlos.

```python
df_grouped = df.groupby('name')[['sex', 'sport']].max()
```

```python
df_ball = df_grouped[df_grouped['sport'].isin(['Basketball', 'Volleyball', 'Football', 'Baseball'])]
```

```python
value_counts_all = df_ball['sport'].value_counts()

value_counts_all
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot()

sns.barplot(x=value_counts_all.index, y=value_counts_all, ax=ax);

ax.set_ylabel('count')
ax.set_title('Conteo de frecuencia de 4 deportes Ol√≠mpicos', fontsize=13, pad=10);
```

¬°Listo! Pero qu√© pasa si ahora adem√°s queremos segmentar nuestros conteos utilizando la variable 'sex'. Esto nos servir√° para saber c√≥mo es que cada g√©nero contribuye al conteo total de atletas.

```python
df_grouped['count'] = 1

df_grouped.head()
```

```python
df_ball_with_count = df_grouped[df_grouped['sport'].isin(['Basketball', 'Volleyball', 'Football', 'Baseball'])]
```

```python
value_counts = df_ball_with_count.groupby(['sport', 'sex'])['count'].sum()

value_counts
```

```python
value_counts = value_counts.unstack(1)

value_counts
```

```python
value_counts = value_counts.fillna(0)
```

```python
plt.style.use('seaborn')

fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot()

plt1 = ax.bar(value_counts.index, value_counts['M'], label='M',
              color=["#7788AA","#4E638E","#2E4372","#152A55"])
plt2 = ax.bar(value_counts.index, value_counts['F'], label='F',
              color=["#FFD0AA","#D4996A","#AA6B39","#804415"])

ax.set_ylabel('count')
ax.set_title('Conteo de frecuencia de 4 deportes Ol√≠mpicos', fontsize=13, pad=15);
plt.legend((plt1[0], plt2[0]), ('Men', 'Women'));
ax.set_ylim(0, 4500);
```

Con esta gr√°fica se vuelve muy evidente la disparidad que existe en casi todos estos deportes. Excepto por volleyball, donde la participaci√≥n de hombres y mujeres es similar, en los dem√°s deportes la participaci√≥n es femenina es mucho menor o incluso nula.
